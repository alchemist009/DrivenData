```python
def funder_process(row):
    if row['funder']=='Government of Tanzania':
        return 'gov'
    elif row['funder'] == 'Danida':
        return 'danida'
    elif row['funder'] == 'Hesawa':
        return 'hesawa'
    elif row['funder'] == 'Rwssp':
        return 'rwssp'
    elif row['funder'] == 'World Bank':
        return 'world_bank'
    else:
        return 'other'
    
df['funder'] = df.apply(lambda row: funder_process(row), axis=1)
```

```python
vals_to_replace = {'functional':2, 'functional needs repair':1, 'non functional' :0}
df['status_group_vals'] = df.status_group.replace(vals_to_replace)
```

```python
piv_table = pd.pivot_table(df, index=['funder', 'status_group'],
                           values='status_group_vals', aggfunc='count')

piv_table
```

```python
df.installer.value_counts()
```

```python
def installer_process(row):
    if row['installer'] == 'Government':
        return 'gov'
    elif row['installer'] == 'DWE':
        return 'dwe'
    elif row['installer'] == 'RWE':
        return 'rwe'
    elif row['installer'] == 'Commu':
        return 'commu'
    elif row['installer'] == 'DANIDA':
        return 'danida'
    else:
        return 'other'
    
df['installer'] = df.apply(lambda row : installer_process(row), axis=1)
```

```python
piv_table = pd.pivot_table(df, index=['funder', 'status_group'], values=
                          'status_group_vals', aggfunc='count')

piv_table
```

```python
def scheme_process(row):
    if row['scheme_management'] == 'VWC':
        return 'vwc'
    elif row['scheme_management']=='WUG':
        return 'wug'
    elif row['scheme_management']=='Water authority':
        return 'wtr_auth'
    elif row['scheme_management']=='WUA':
        return 'wua'
    else:
        return 'other'

df['scheme_management'] = df.apply(lambda row : scheme_process(row), axis=1)
```

```python
def construction_process(row):
    if row['construction_year'] >= 1960 and row['construction_year'] < 1970:
        return '60s'
    elif row['construction_year'] >= 1970 and row['construction_year'] < 1980:
        return '70s'
    elif row['construction_year'] >= 1980 and row['construction_year'] < 1990:
        return '80s'
    elif row['construction_year'] >= 1990 and row['construction_year'] < 2000:
        return '00s'
    elif row['construction_year'] >= 2000 and row['construction_year'] < 2010:
        return '10s'
    else:
        return 'Unknown'
    
df['construction_year'] = df.apply(lambda row : construction_process(row), axis=1)
```
